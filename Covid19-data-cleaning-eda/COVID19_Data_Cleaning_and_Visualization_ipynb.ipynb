# **Clean DATE_DIED Column**

```python
df['DATE_DIED'].value_counts()
```

```python
df['Death'] = np.where(df['DATE_DIED']=='9999-99-99',
                       'Alive','Dead')
```

```python
df.drop('DATE_DIED', axis=1, inplace=True)
df.head()
```

```python
df['PNEUMONIA'].unique()
```

# **Clean PNEUMONIA Column**

```python
df['PNEUMONIA'] = df['PNEUMONIA'].replace(99, np.nan)
```

```python
df['PNEUMONIA'].fillna(df['PNEUMONIA'].mode()[0], inplace=True)
```

```python
df['PNEUMONIA'].value_counts()
```

```python
df['PREGNANT'].value_counts()
```

# **Clean PREGNANT Column**

```python
df['PREGNANT'].replace(4, np.nan, inplace=True)
```

# **Fill Missing Values**

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# **Visualization part**

```python
plt.figure(figsize=(16,10))

corr = df.corr(numeric_only=True)

sns.heatmap(corr,
            cmap='coolwarm',
            annot=False)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x=df['AGE'])

plt.title("Age Outliers")
plt.show()
```

```python
original_df = df.copy()
```

```python
Q1 = df['AGE'].quantile(0.25)
Q3 = df['AGE'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df = df[(df['AGE'] >= lower) &
        (df['AGE'] <= upper)]

df.shape
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x=df['AGE'])

plt.title("Age Distribution After Outlier Removal")
plt.xlabel("Age")

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))

# Before Removal
sns.boxplot(x=original_df['AGE'],
            ax=ax[0],
            color='skyblue')

ax[0].set_title("Before Outlier Removal")

# After Removal
sns.boxplot(x=df['AGE'],
            ax=ax[1],
            color='lightgreen')

ax[1].set_title("After Outlier Removal")

plt.tight_layout()
plt.show()
```
